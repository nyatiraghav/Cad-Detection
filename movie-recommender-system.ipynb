```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('/kaggle/input/tmdb-5000-movies-data/tmdb_5000_movies.csv')
credits = pd.read_csv('/kaggle/input/tmdb-5000-movies-data/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits)
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies = movies[['genres', 'id', 'keywords', 'overview', 'cast', 'crew', 'title']]
```

```python
movies.head()
```

```python
#checking if there are any null values in the selected data
movies.isnull().sum()
```

```python
#dropping all the null values
movies.dropna(inplace = True)
```

```python
#checking if there are any duplicated data
movies.duplicated().sum()
```

```python
#to convert string of list into a list we import ast module and use literal_eval function from it 
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
        
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#selecting only top three members of cast 
def convert3(obj):
    L = []
    c = 0
    for i in ast.literal_eval(obj):
        if c!=3:
            L.append(i['name'])
            c=+1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
#we only seek director from crew
def fetchDirector(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetchDirector)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew']
```

```python
movies = movies[['id', 'title', 'tags']]
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies['tags'] = movies['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)  
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
feature_names = cv.get_feature_names_out()
print(feature_names)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0] 
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies.iloc[i[0]].title)
```
